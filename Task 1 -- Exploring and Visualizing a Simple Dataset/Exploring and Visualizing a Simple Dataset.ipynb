```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

# STEP 1: LOADING THE DATA
df = sns.load_dataset('iris')

# STEP 2: SUMMARIZING THE DATA
print("1. DATASET SHAPE (Rows, Columns):")
print(df.shape)

print("\n2. NAMES OF THE COLUMNS:")
print(df.columns)

print("\n3. FIRST 5 ROWS OF DATA:")
print(df.head())

# STEP 3: VISUALIZING THE DATA

# A) SCATTER PLOT
# This helps us see how Sepal Length relates to Sepal Width
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')
plt.title('Relationship between Sepal Length and Width')
plt.show()

# B) HISTOGRAM
# This shows us the distribution (spread) of Petal Length
plt.figure(figsize=(8, 5))
sns.histplot(df['petal_length'], color='skyblue', kde=True)
plt.title('Distribution of Petal Length')
plt.show()

# C) BOX PLOT
# This helps us see the average Petal Width for each flower species
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='species', y='petal_width')
plt.title('Petal Width by Flower Species')
plt.show()
```

```python
# Save the data to a CSV file on your computer
df.to_csv('iris_dataset.csv', index=False)
print("Dataset saved as iris_dataset.csv!")
```
